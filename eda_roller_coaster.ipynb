```python
df=df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
       #'Type', 
       'Manufacturer',
        # 'Height restriction', 'Model', 'Height',
       #'Inversions',
        # 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
       #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 'height_value', 'height_unit', 'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.dtypes
```

```python
df['Opening_Date']=pd.to_datetime(df['opening_date_clean'])
```

```python
df=df.rename(columns={'coaster_name':'Coaster_Name',
          'year_introduced':'Year_Introduced',
          'latitude':'Latitude', 
          'longitude':'Longitude', 
          'speed_mph':'Speed_Mph',
          'height_value':'Height_Value', 
          'height_unit':'Height_Unit', 
          'height_ft':'Height_ft', 
          'Inversions_clean':'Inversions_Clean',
          'Gforce_clean':'Gforce_Clean'
          })
```

```python
df=df.drop(columns={'opening_date_clean'},axis=1)
```

```python
df=df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy()
```

```python
df['Year_Introduced'].value_counts
```

```python
ax=df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title="top 10 year Roller Coaster Introduced")
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
ax=df['Speed_Mph'].plot(kind='hist',
                     bins=20,
                     title="Coaster speed(Mph)")
ax.set_xlabel('Coaster Speed')
```

```python
ax=df['Speed_Mph'].plot(kind='kde',
                     title="Coaster speed(Mph)")
ax.set_xlabel('Coaster Speed')
```

```python
df.plot(kind="scatter",
        x='Speed_Mph',
        y='Height_ft',
        title='Coaster Speed vs Height')
plt.show()
```

```python
ax= sns.scatterplot(x='Speed_Mph',
                y='Height_ft',
                hue='Year_Introduced',
                data=df)
ax.set_title('Coaster Speed vs Height')
plt.show()
```

```python
sns.pairplot(df,vars=['Year_Introduced','Speed_Mph','Height_ft', 'Inversions_Clean', 'Gforce_Clean',
       'Opening_Date'],
       hue='Type_Main')
plt.show()
```

```python
df_corr=df[['Year_Introduced','Speed_Mph','Height_ft', 'Inversions_Clean', 'Gforce_Clean',
       'Opening_Date']] \
        .dropna()\
        .corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```

```python
ax= df.query('Location != "Other"') \
    .groupby('Location')['Speed_Mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12,3), title='Average Coaste Speed by Locations')
ax.set_xlabel('Average Speed')
plt.show()
```
